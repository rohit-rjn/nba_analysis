### Logistic Regression for Win Predictibility
In this report I am trying to predict win probability of a team in a given match<br>
To build a Good Logistic Regression Model, I am trying to build a good feature set<br>
The feature set containes: <br>
1) Home Team vs Visiter Team <br>
2) Perfomrance by team in last season (Win/Losses) <br>
3) Performance in this season till now (Win/Losses) <br>
4) Points Used Till Mid-Time <br>

The abvoe feature set is used, as it available pre-match, and we are trying to predict the win probability in two steps <br>
1) Before the Start of Match based on first 3 stats, <br>
2) After Half Time by using 2 QTRs data, and also transforming into mean Diff by first half <br>

### Adding Data to DataFrame for Creating Feature Set
1) Home/Visitor Team to each row (HOME_TEAM=1 or HOME_TEAM=0)<br>

3) Last Season Win-Loss PCT- which is split into 2 parts- Last 20 games and all the Games <br>

```python
# Adding Column Win/Loss by comparing the PTS for each game between 2 teams
# Adding Leading in First Half into Data Frame
# Adding Points Differential (Difference in Points at Half Time)
df['WIN'] = 0
df['LEAD_FH'] = 0
df['HALF_PTS_DIFF'] = 0
for i in range(0,len(df),2):
    if df['PTS'].iloc[i] > df['PTS'].iloc[i+1]: 
        df['WIN'].iloc[i] = 1
    elif df['PTS'].iloc[i] < df['PTS'].iloc[i+1]: 
        df['WIN'].iloc[i+1] = 1
    elif df['PTS'].iloc[i] == df['PTS'].iloc[i+1]:
        df['WIN'].iloc[i] = 2
        df['WIN'].iloc[i+1] =2 
        
    if df['PTS_FHALF'].iloc[i] > df['PTS_FHALF'].iloc[i+1]: 
        df['LEAD_FH'].iloc[i] = 1
    elif df['PTS_FHALF'].iloc[i] < df['PTS_FHALF'].iloc[i+1]: 
        df['LEAD_FH'].iloc[i+1] = 1
    elif df['PTS_FHALF'].iloc[i] == df['PTS_FHALF'].iloc[i+1]:
        df['LEAD_FH'].iloc[i] = 2
        df['LEAD_FH'].iloc[i+1] =2 
        
    df['HALF_PTS_DIFF'].iloc[i] = (df['PTS_FHALF'].iloc[i] - df['PTS_FHALF'].iloc[i+1])
    df['HALF_PTS_DIFF'].iloc[i+1] = -(df['HALF_PTS_DIFF'].iloc[i])
```

### Part 1: Logistic Regression Using Only Previous Data Available (No Quater Points)
1) We need to convert the output to 0-1 (Win-Loss), I am converting all Tied to 0 <br>
2) Removing First 100 Rows, as the data for Win% this season would not be ideal for Classification <br>

```python
# For Logistic Regression - We n
df_lr = df.loc[100:]
df_lr = df_lr.replace({'WIN':2},0)
X = df_lr[['HOME_TEAM','LAST_20_WIN_PCT','PAST_WIN_PCT','2018_WIN_LOSS_PCT']]

y = df_lr['WIN']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
# Precision and Recall of the above classification
print(classification_report(y_test, y_pred))
```

### Team Statistics for Season 2018-2019
1) Transform data from Game/Team to Team Summery <br>
2) Sort the Team by most number of wins for the Season <br>
3) Find the Mean of QTR points and Other Factors of the Game <br>


```python
df_factor = pd.read_csv('2018-2019_factor.csv',index_col=0)
```

```python
#Transform the Data from Team/Game to Team specific for Number of Wins and Number of Losses

columns = ['WIN','PTS_QTR1','PTS_QTR2','PTS_QTR3','PTS_QTR4','PTS_OT1','PTS_FHALF','PTS_SHALF','PTS']
columns_factor = ['OPP_OREB_PCT','OPP_TOV_PCT','OPP_FTA_RATE','OPP_EFG_PCT','OREB_PCT','TM_TOV_PCT','FTA_RATE','EFG_PCT']
for column in columns:
    df_team[column] = 0
for column in columns_factor:
    df_team[column] = 0
df_team['NO_GAMES'] = 0
for i in range(len(df_team)):
    df_sum = df[df.TEAM_ID == df_team['TEAM_ID'].iloc[i]].sum()
    df_mean_factor =  df_factor[df_factor.TEAM_ID == df_team['TEAM_ID'].iloc[i]].mean()
    for column in columns:
        df_team[column].iloc[i] = df_sum[column]
    for column in columns_factor:
        df_team[column].iloc[i] = df_mean_factor[column]
        
    df_team['NO_GAMES'].iloc[i] = df[df.TEAM_ID == df_team['TEAM_ID'].iloc[i]].GAME_ID.count()
    
```

```python
# Assuming not won as Lossing- (Not considering the Tied Games)
df_team['LOSS']     = df_team['NO_GAMES'] - df_team['WIN']
df_team['WIN_PCT']  = df_team['WIN']/df_team['NO_GAMES']
df_team['LOSS_PCT'] = df_team['LOSS']/df_team['NO_GAMES']

# Adding Mean of Points Scored in Half time and Total Points
df_team['PTS_FHALF_MEAN'] = df_team['PTS_FHALF']/df_team['NO_GAMES']
df_team['PTS_MEAN'] = df_team['PTS']/df_team['NO_GAMES']
```

```python
# Simple Table For showing the Ranking of Each Team at the end of Season 2018-2019
df_team = df_team.sort_values(by='WIN', ascending=False)
#df_team = df_team.reset_index()
df_team_abbr = df_team[['TEAM_ID','ABBREVIATION','WIN','WIN_PCT','LOSS','LOSS_PCT','PTS_FHALF_MEAN','PTS_MEAN']]
pd.set_option('expand_frame_repr', True)
df_team_abbr
```
